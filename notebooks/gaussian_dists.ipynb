```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.style as style
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource
from matplotlib import cm

import seaborn as sns
from scipy.stats import norm, multivariate_normal
```

```python
x = np.linspace(norm.ppf(0.001),norm.ppf(0.999),1000)
y = np.linspace(norm.ppf(0.001),norm.ppf(0.999),1000)
xi, yi = np.meshgrid(x,y)
```

```python
fig, ax = plt.subplots(1,1,figsize=(8,5))

plt.plot(x,norm.pdf(x))
plt.xlabel(r'$X$')
plt.ylabel('Probability Density')
plt.tick_params(axis='y',which='both',left=False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.savefig('figures/gaussian_pdf.png',dpi=200)
plt.show()
```

```python
x = np.linspace(norm.ppf(0.01),norm.ppf(0.99),1000)
y = np.linspace(norm.ppf(0.01),norm.ppf(0.99),1000)
xi, yi = np.meshgrid(x,y)
sampling = np.dstack((xi,yi))
rv = multivariate_normal([0, 0],[[0.5, 0.3], [0.3, 0.8]])
```

```python
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111,projection="3d")


ax.set_zlim(bottom=-0.3,top=0.25)

ax.plot_surface(sampling[:,:,0],sampling[:,:,1],rv.pdf(sampling),
            cmap='inferno_r',rcount=25,ccount=25,lw=0.5,edgecolors='k',
               lightsource=True)

ax.contour(sampling[:,:,0],sampling[:,:,1],rv.pdf(sampling),10,
          cmap='inferno_r', offset=-0.3,linestyles='solid')

ax.set_xlabel(r'$X_1$')
ax.set_ylabel(r'$X_2$')
ax.set_zlabel('Probability Density')

plt.tight_layout()
plt.savefig('figures/2d_gauss.png',dpi=300)
plt.show()
```
